```python
# best threshold identification :: 

import os
import numpy as np
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

model = tf.keras.models.load_model(r"C:\Users\Devansh\Desktop\Project Backups\Model2\LungAbnormalityDetectorTwoModel.h5", compile=False)

test_dir = r"C:\Users\Devansh\Desktop\Project Backups\LungAbnormalityDetectorDataset(PreprocessedAugmented)\Test"
IMG_SIZE = (256, 256)
BATCH_SIZE = 32

test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    test_dir,
    label_mode='binary',
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)


from sklearn.metrics import precision_score, recall_score, f1_score

# Store raw predicted probabilities for each sample
all_probs = []

y_true = []
for batch_images, batch_labels in test_ds:
    probs = model.predict(batch_images).flatten()
    all_probs.extend(probs)
    y_true.extend(batch_labels.numpy().astype(int))

y_true = np.array(y_true)
all_probs = np.array(all_probs)

# Test many thresholds
thresholds = np.arange(0.01, 1.00, 0.01)
precision_list = []
recall_list = []
f1_list = []

for t in thresholds:
    preds = (all_probs > t).astype(int)
    precision = precision_score(y_true, preds, zero_division=0)
    recall = recall_score(y_true, preds, zero_division=0)
    f1 = f1_score(y_true, preds, zero_division=0)
    precision_list.append(precision)
    recall_list.append(recall)
    f1_list.append(f1)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(thresholds, precision_list, label='Precision', marker='.')
plt.plot(thresholds, recall_list, label='Recall', marker='.')
plt.plot(thresholds, f1_list, label='F1-Score', marker='.')
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.title("Precision, Recall & F1 vs Threshold")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print the best F1 threshold
best_idx = np.argmax(f1_list)
print(f"\n✅ Best Threshold = {thresholds[best_idx]:.2f} with F1-Score = {f1_list[best_idx]:.4f}")
```

```python
# model testing ::

import os
import numpy as np
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# === USER SETTINGS ===
model_path = r"C:\Users\Devansh\Desktop\Project Backups\Model2\LungAbnormalityDetectorTwoModel.h5"  # Your trained model
test_dir = r"C:\Users\Devansh\Desktop\Project Backups\LungAbnormalityDetectorDataset(PreprocessedAugmented)\Test" 

IMG_SIZE = (256, 256)
BATCH_SIZE = 32
threshold = 0.21  # Based on training
class_names = ["Abnormal", "Normal"]

# === Load Model ===
model = tf.keras.models.load_model(model_path, compile=False)

# === Load Dataset ===
test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    test_dir,
    label_mode='binary',
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

# === Predict & Evaluate ===
y_true, y_pred = [], []

# Get file paths in the same order as test_ds batches
all_file_paths = test_ds.file_paths  # tf >= 2.5
# If not available, use:
# all_file_paths = [f for subdir, _, files in os.walk(test_dir)
#                   for f in sorted(files)
#                   if f.lower().endswith(('.png','.jpg','.jpeg'))]

idx = 0
for batch_images, batch_labels in test_ds:
    probs = model.predict(batch_images).flatten()
    preds = (probs > threshold).astype(int)

    batch_size = len(batch_labels)
    for i in range(batch_size):
        img_path = all_file_paths[idx]
        bin_pred = preds[i]
        print(f"{os.path.basename(img_path):<30} → Predicted: {class_names[bin_pred]}")
        idx += 1

    y_true.extend(batch_labels.numpy().astype(int))
    y_pred.extend(preds)

# === Evaluation Metrics ===
print("\n📊 Classification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title(f"Confusion Matrix (Threshold = {threshold})")
plt.tight_layout()
plt.show()
```
