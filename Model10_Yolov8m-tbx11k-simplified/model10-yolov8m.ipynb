```python
!pip install -q ultralytics

from ultralytics import YOLO
import os
import shutil
import matplotlib.pyplot as plt
```

```python
# ==== Define paths ====
DATA_YAML_PATH = '/kaggle/input/yolo-tbx11k-simplified/yolo-tbx11k-simplified/data.yaml'
WORK_DIR = '/kaggle/working'  # Model will be saved here
MODEL_NAME = 'yolov8m'         # Using the medium model

# ==== Load model ====
model = YOLO(f'{MODEL_NAME}.pt')  # Pretrained YOLOv8m weights

# ==== Train model ====
model.train(
    data=DATA_YAML_PATH,
    epochs=200,
    imgsz=512,
    batch=16,
    patience=20,
    device=0,  # Auto-selects GPU
    project=WORK_DIR,
    name='tb_detector_yolov8m',
    exist_ok=True,
    save=True,
    verbose=True
)
```
